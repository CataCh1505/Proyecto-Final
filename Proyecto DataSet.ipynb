<h5 style='color:#f2c08b;'>5.1 Configuración para los gráficos</h5>

```python
#Remove all axis
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
Pregunta 1: ¿Cuál es el ingreso anual de las propiedades vendidas?
```

```python
gf = df_dataSet.groupby(['Sale-Amount']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('Serial-Number', ascending = False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,6.9), dpi=100)

eje_x = gf.index.tolist()
data = gf['Serial-Number'].tolist()


bar_colors = ['#12233d','#667a75','#1b4022']
ax.barh(eje_x, data, color=bar_colors)

 

#--------------------------------------------------------------------------------------------------------------------#
#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO
title = "Grafico 1: Ingreso anual"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Arial")

 
#FORMATO DEL SUBTITULO
subTitle = "Ingreso anual de las propiedades vendidas"
ax.set_title(subTitle,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

#FORMATO DE LOS EJES
etiqueta = "Ingreso anual"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

#CONFIGURACION DE COORDENADAS DE LA ETIQUETA DEL EJE X
ax.xaxis.set_label_coords(x=0,
                          y=-0.07,
                          transform=ax.transAxes)

#CONFIGURACION EL TAMANO DEL TEXTO DEL EJE Y
ax.tick_params(axis='both',
               labelsize=10)

#CONFIGURACION LA UBICACION DEL GRAFICO
plt.subplots_adjust(top=0.80)
plt.show()
```

```python
Pregunta 2: ¿Qué tipo de propiedas tiene mayor facilidad de venta?
```

```python
gf = df_dataSet.groupby(['Property-Type']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('Serial-Number', ascending = False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,6.9), dpi=90)

eje_x = gf.index.tolist()
data = gf['Serial-Number'].tolist()

bar_colors = ['#82877c','#2b2e4e','#7ea7e5']
ax.pie(data, labels=eje_x,colors=bar_colors, autopct='%1.1f%%')
 
#--------------------------------------------------------------------------------------------------------------------#
#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#
#--------------------------------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO
title = "Grafico 1: Propiedas mayor venta"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Arial")

 
#FORMATO DEL SUBTITULO
subTitle = "Propiedas tiene mayor facilidad de venta"
ax.set_title(subTitle,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")


#CONFIGURACION LA UBICACION DEL GRAFICO
plt.subplots_adjust(top=0.80)
plt.show()
```
